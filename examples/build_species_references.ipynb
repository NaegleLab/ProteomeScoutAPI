```python
from pathlib import Path
from proteomeScoutAPI import config, SpeciesReferenceDataset

# Use this if building a specific species dataste from a new local dataset file. 
dataset_root = Path("/path/to/dataset")  # <-- UPDATE THIS TO YOUR LOCAL PATH
expected_data_file = dataset_root / "ProteomeScout_Dataset" / "data.tsv"
if not expected_data_file.exists():
    raise FileNotFoundError(f"Could not find local dataset file: {expected_data_file}")

out_directory = "../Data/species_reference_datasets"
original_dataset_dir = config.DATASET_DIR

try:
    config.DATASET_DIR = str(dataset_root)
    # Keep this False so version mismatch checks do not auto-download.
    b = SpeciesReferenceDataset(update=False, version=8)
    out = b.write_all_species_reference_datasets(out_directory)
    print("Wrote", len(out), "files to", out_directory)
finally:
    config.DATASET_DIR = original_dataset_dir
```
